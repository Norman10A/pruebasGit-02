```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar estilo
sns.set(style="whitegrid")

# Dataset de ejemplo
df = sns.load_dataset("iris")

# Gráfico
grafico = sns.barplot(
    x="species",
    y="petal_length",
    data=df,
    errorbar=None
)

grafico.set_title("Promedio de Longitud de Pétalos por Especie", fontsize=14)
plt.show()
```

```python
import pandas as pd

class DataAnalyzer:
    def __init__(self):
        self.data = None

    def cargar_datos(self, ruta):
        """Carga un archivo CSV."""
        try:
            self.data = pd.read_csv(ruta, encoding="utf-8")
            print("--- Datos cargados correctamente ---")
        except FileNotFoundError:
            print(f"Error: El archivo '{ruta}' no existe.")
        except pd.errors.ParserError as e:
            print(f"Error al leer el CSV: {e}")

    def limpiar_datos(self):
        """Elimina duplicados y filas con valores nulos."""
        if self.data is not None:
            self.data.drop_duplicates(inplace=True)
            self.data.dropna(inplace=True)
            print("Datos limpiados correctamente.")
        else:
            print("No hay datos cargados para limpiar.")

    def resumen(self):
        """Muestra estadísticas y valores nulos."""
        if self.data is not None:
            print("Resumen estadístico:\n")
            print(self.data.describe())
            print("\nValores nulos por columna:\n")
            print(self.data.isnull().sum())
        else:
            print("No hay datos cargados para resumir.")

    def mostrar_datos(self, n=5):
        """Muestra las primeras n filas."""
        if self.data is not None:
            print(self.data.head(n))
        else:
            print("No hay datos cargados para mostrar.")

```

```python
analizador = DataAnalyzer()
analizador.cargar_datos("datos.csv")
analizador.mostrar_datos()
analizador.limpiar_datos()
analizador.resumen()
```
